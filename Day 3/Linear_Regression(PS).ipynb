```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/SmartPickAI_dataset.csv')
df
```

```python
X = df[['rating']]
y = df[['price']]
```

```python
model = linear_model.LinearRegression()
model.fit(X,y)
```

```python
print("Predicted Value:",model.predict([[3.6]]))
```
